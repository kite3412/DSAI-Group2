```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
tit1 = soup.find("h1", "gepq850 eihuid4z i7p5ej0 i7p5ejl _18ybopc4 i7p5ejs i7p5ej21")
```

```python
tit = [tit1.text]
```

```python
tit
```

```python
sal1 = soup.find("span", {"data-automation": "job-detail-salary"})
```

```python
sal = [sal1.text]
```

```python
sal
```

```python
des = soup.find_all("ul")
```

```python
res = [res.text for res in des[0].find_all("li")]
```

```python
res
```

```python
req = [req.text for req in des[1].find_all("li")]
```

```python
req
```

```python
dic = {
    "Title": tit,
    "Salary": sal,
    "Responsibilities": res,
    "Requirements": req
}
```

```python
df = pd.DataFrame({
    "Title": dic["Title"],
    "Salary": dic["Salary"],
    "Responsibilities": [dic["Responsibilities"]],
    "Requirements": [dic["Requirements"]]
})
```
